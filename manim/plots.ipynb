```python

import matplotlib.pyplot as plt
```

```python
contrast_color = "#e63946"

other_colors = ["#f1faee", "#a8dadc", "#457b9d", "#1d3557",]
```

```python
plt.bar(0, 76, color=contrast_color, edgecolor=other_colors[-1])
plt.bar(x=[1, 2, 3], height=[85, 81, 95], color=other_colors[2], edgecolor=other_colors[-1], label="MLP")
bar = plt.bar(x=[1, 2, 3], height=[80, 80, 89], color=other_colors[0], edgecolor=other_colors[-1], label="KNN", linestyle="--", alpha=0.7)
plt.xticks(range(4), ["Base", "DINO", "CodeBERT", "UnixCoder"])
plt.ylim(70, 100)
for i, v in enumerate([76, 85, 81, 95]):
    plt.text(i, v + 1, f"{v}%", horizontalalignment='center')
plt.legend(loc='upper left')
plt.ylabel("Accuracy")
plt.title("Task classification")
plt.show()
```
